```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
!wget https://a.espncdn.com/combiner/i?img=/photo/2021/0504/r849562_1296x729_16-9.jpg
```

```python
img=cv2.imread("/content/i.jfif")
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```python
np.array(img).shape
```

```python
img_pixels=img.reshape((-1,3))
img_pixels=np.float32(img_pixels)
```

```python
img_pixels.shape
```

```python
fig=plt.figure(figsize=(10,10))
plt.title("Image segmentation with different no. of clusters")
for i in range(1,17):
  k = i
  criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.85)
  retval, labels, centers = cv2.kmeans(img_pixels, k, None, criteria,10, cv2.KMEANS_RANDOM_CENTERS)
  centers = np.uint8(centers)
  segmented_data = centers[labels.flatten()]
  segmented_image = segmented_data.reshape((img.shape))
  ax=fig.add_subplot(4,4,i)
  ax.set_title("k="+str(i))
  ax.imshow(segmented_image)
```
